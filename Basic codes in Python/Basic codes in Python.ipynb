## Tuple

**Definition**: In mathematics, a tuple is a finite ordered list (sequence) of elements. An n-tuple is a sequence (or ordered list) of n elements, where n is a non-negative integer. In Python, Tuple is a collection of objects separated by commas. In some ways, a tuple is similar to a list in terms of indexing, nested objects, and repetition but a tuple is immutable, unlike lists which are mutable.

````python
a = (1, 2, 3, 4, 5)
b = ("a", "b", "c", "d")

print(f"a = {a}")
print(f"len(a) = {len(a)}")
print(f"sum(a) = {sum(a)}")
print(f"min(a) = {min(a)}")
print(f"max(a) = {max(a)} \n")

print(f"b = {b}")
print(f"b[0] = {b[0]}")
print(f"b[-1] = {b[-1]}")
print(f"b[1:4] = {b[1:4]}")
````

## Sets

Sets are made to represent the formal mathematical definition of sets. Namely, unordered collections of elements without repetition of elements. So it is important to remember:  
- Sets are unordered.  
- Set elements are unique. Duplicate elements are not allowed.  
- A set itself may be modified, but the elements contained in the set must be of an immutable type.  

````python
# sets are unordered
set_one = {"a", "b", "c", "d"}
set_two = set(["c", "d", "e"])

print(f"set_one = {set_one}")
print(f"set_two = {set_two}")
````

````python
# duplicate elements are not allowed
x = {"a", "a", "b", "c", "c", "d"}
print(f"x = {x}")
````

````python
## we can use set to remove duplicate elements
data = [100, 90, 100, 30, 30, 40, 30]
print(f"data = {data}")
print(f"unique data entries = {list(set(data))}")
````

````python
# Set Operators
A = {"a", "b", "c", "d", "e"}
B = {"c", "d", "e", "f", "g"}

print(f"A = {A}")
print(f"B = {B} \n")
print("---------------- Set Operators ---------------- \n")
print(f"Set Union: A or B = {A.union(B)} \n")
print(f"Set Intersection: A and B = {A.intersection(B)} \n")
print(f"Set Difference: A - B = {A.difference(B)} \n")
print(f"Set Symmetric Difference: A (+) B = {A.symmetric_difference(B)} \n")
````

## Polynomial example

**Problem Statement**  
  
Suppose we are dealing with polynomial functions of the form:   
$P(x)=a_nx^n+a_{n-1}x^{n-1}+ \cdots +a_1x+a_0$  


Let us establish a polynomial class which can print, calculate based on the given x value and do the basic calculation for two functions.

````python
# Create custom Polynomial object

class Polynomial(object):                        # define the class

    def __init__(self, coef, name="f(x)"):          # initialize the polynomial
        self.coef = coef                         # list of coefficients
        self.degree = len(coef) - 1              # degree of the polynomial
        self.name = name                         # name of the polynomial
    
    def __call__(self, x):                       # evaluate the polynomial at x
        y = 0
        for i in range(self.degree + 1):         # loop through all coefficients
            y += self.coef[i] * x ** i           # add the coefficient times x to the power
        return y                                 # return the value of the polynomial
    
    def __str__(self):                           # print the polynomial
        s = ""                                   # initialize the string
        for i in range(self.degree + 1):         # loop through all coefficients
            s += f"{self.coef[i]}x^{i} + "       # add the coefficient and the power
        return s[:-3]                            # remove the last " + "
    
    def __add__(self, other):                    # add two polynomials
        if self.degree > other.degree:           # if self has a higher degree
            coef = self.coef[:]                  # copy the coefficients
            for i in range(other.degree + 1):    # loop through all coefficients
                coef[i] += other.coef[i]         # add the coefficients
        else:                                    # if other has a higher degree
            coef = other.coef[:]                 # copy the coefficients
            for i in range(self.degree + 1):     # loop through all coefficients
                coef[i] += self.coef[i]          # add the coefficients
        return Polynomial(coef)                  # return a new Polynomial object
    
    def __sub__(self, other):                    # subtract two polynomials
        if self.degree > other.degree:           # if self has a higher degree
            coef = self.coef[:]                  # copy the coefficients
            for i in range(other.degree + 1):    # loop through all coefficients
                coef[i] -= other.coef[i]         # subtract the coefficients
        else:                                    # if other has a higher degree
            coef = [-x for x in other.coef]      # copy the coefficients and negate them
            for i in range(self.degree + 1):     # loop through all coefficients
                coef[i] += self.coef[i]          # add the coefficients
        return Polynomial(coef)                  # return a new Polynomial object
    
    def __mul__(self, other):                    # multiply two polynomials
        coef = [0] * (self.degree + other.degree + 1) # initialize the coefficients
        for i in range(self.degree + 1):         # loop through all coefficients
            for j in range(other.degree + 1):    # loop through all coefficients
                coef[i + j] += self.coef[i] * other.coef[j] # multiply the coefficients
        return Polynomial(coef)                  # return a new Polynomial object
   
````

````python
# I want to create a polynomial object, f(x) = 1 + 2x + 3x^2 + 4x^3 + 5x^4

coef=[1, 2, 3, 4, 5]
f = Polynomial(coef, "f(x)")

# I want to see the function
print(f"f(x)={f})")

# I want to evaluate the function at x = 2
print(f"f(x=2)={f(2)}")

# I want to add two polynomials
g = Polynomial([1, 2, 3, 4, 5], "g(x)")
print(f"g(x)={g})")
h = f + g
print(f"h(x)=f(x)+g(x)={h}")

# I want to subtract two polynomials
h = f - g
print(f"h(x)=f(x)-g(x)={h}")

# I want to multiply two polynomials
h = f * g
print(f"h(x)=f(x)*g(x)={h}")
````
